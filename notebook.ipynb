### Analisis Fitur Kategorikal vs Status

```python
category_maps = {
    'Marital_status': {
        1: 'Belum Menikah',
        2: 'Menikah',
        3: 'Janda/Duda',
        4: 'Cerai',
        5: 'Union De Facto',
        6: 'Pisah Secara Hukum'
    },
    'Daytime_evening_attendance': {
        1: 'Siang',
        0: 'Malam'
    },
    'Debtor': {
        1: 'Memiliki Utang',
        0: 'Tidak Memiliki Utang'
    },
    'Tuition_fees_up_to_date': {
        1: 'Tertib Pembayaran',
        0: 'Tidak Tertib Pembayaran'
    },
    'Gender': {
        1: 'Pria',
        0: 'Wanita'
    },
    'Scholarship_holder': {
        1: 'Penerima Beasiswa',
        0: 'Bukan Penerima Beasiswa'
    },
    'International': {
        1: 'Mahasiswa Internasional',
        0: 'Mahasiswa Lokal'
    }
}

feature_titles = {
    'Marital_status': 'Status Perkawinan',
    'Daytime_evening_attendance': 'Waktu Kuliah (Siang/Malam)',
    'Debtor': 'Status Peminjaman',
    'Tuition_fees_up_to_date': 'Status Pembayaran Uang Kuliah',
    'Gender': 'Jenis Kelamin',
    'Scholarship_holder': 'Penerima Beasiswa',
    'International': 'Mahasiswa Internasional'
}

def map_categories(df, category_maps):
    df_mapped = df.copy()
    for feature, mapping in category_maps.items():
        if feature in df_mapped.columns:
            df_mapped[feature] = df_mapped[feature].map(mapping)
    return df_mapped

df_mapped = map_categories(df, category_maps)

cat_features = ['Marital_status', 'Daytime_evening_attendance', 'Debtor',
                'Tuition_fees_up_to_date', 'Gender', 'Scholarship_holder', 'International']

color_map = {'Graduate': '#1f77b4', 'Enrolled': '#2ca02c', 'Dropout': '#ff7f0e'}

for feature in cat_features:
    df_feature = df_mapped.groupby([feature, 'Status'], observed=False).size().reset_index(name='count')
    df_feature['percentage'] = df_feature.groupby(feature, observed=False)['count'].transform(lambda x: x / x.sum() * 100)
    
    title = f'Distribusi Status Berdasarkan {feature_titles[feature]}'
    
    fig = px.bar(df_feature, x=feature, y='percentage', color='Status',
                 barmode='group',
                 title=title,
                 color_discrete_map=color_map)
    
    fig.update_layout(
        bargap=0.2,
        plot_bgcolor='white',
        xaxis=dict(
            showline=True, 
            linewidth=2, 
            linecolor='black', 
            title='Kategori'
        ),
        yaxis=dict(
            showline=True, 
            linewidth=2, 
            linecolor='black', 
            showgrid=True, 
            gridcolor='lightgrey', 
            gridwidth=0.5, 
            title='Persentase'
        ),
    )

    fig.show()
```

## Data Preparation / Preprocessing

**Menghapus data dengan status "Enrolled"**

```python
df_model = df[df['Status'].isin(['Graduate', 'Dropout'])].copy()
```

**Encode label dan fitur kategorikal**

```python
df_model['Status'] = df_model['Status'].map({'Dropout': 1, 'Graduate': 0})

categorical_cols = df_model.select_dtypes(include='object').columns
le = LabelEncoder()
for col in categorical_cols:
    df_model[col] = le.fit_transform(df_model[col])

joblib.dump(le, 'model/label_encoder.joblib')
```

**Feature Engineering**  
Academic_consistency: selisih antara nilai semester 1 & 2

```python
df_model['Academic_consistency'] = abs(df_model['Curricular_units_1st_sem_grade'] - df_model['Curricular_units_2nd_sem_grade'])
```

Total_failed_units: total mata kuliah yang tidak lulus

```python
df_model['Total_failed_units'] = (df_model['Curricular_units_1st_sem_enrolled'] - df_model['Curricular_units_1st_sem_approved']) + \
    (df_model['Curricular_units_2nd_sem_enrolled'] - df_model['Curricular_units_2nd_sem_approved'])
```

Total_evaluations: total evaluasi semester 1 + 2

```python
df_model['Total_evaluations'] = df_model['Curricular_units_1st_sem_evaluations'] + df_model['Curricular_units_2nd_sem_evaluations']

model_columns = df_model.drop('Status', axis=1).columns.tolist()
joblib.dump(model_columns, 'model/model_columns.joblib')
```

**Memisahkan data menjadi data latih dan data uji**

```python
X = df_model.drop('Status', axis=1)
y = df_model['Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

**Feature Scaling**

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

joblib.dump(scaler, 'model/scaler.joblib')
```

## Modeling

### Logistic Regression

```python
logreg = LogisticRegression(solver='saga', max_iter=10000, random_state=42)
logreg.fit(X_train, y_train)

joblib.dump(logreg, 'model/logreg_model.joblib')
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

joblib.dump(rf, 'model/rf_model.joblib')
```

## Evaluation

**Classification Report**

```python
y_pred_log = logreg.predict(X_test)
y_prob_log = logreg.predict_proba(X_test)[:, 1]
print("Logistic Regression Report:\n", classification_report(y_test, y_pred_log))
print("ROC AUC (Logistic Regression):", roc_auc_score(y_test, y_prob_log))

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]
print("\nRandom Forest Report:\n", classification_report(y_test, y_pred_rf))
print("ROC AUC (Random Forest):", roc_auc_score(y_test, y_prob_rf))
```
